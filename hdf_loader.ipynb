## How to load data from HDF files

```python
import dolfin
```

```python
# load mesh
mesh = dolfin.Mesh("code/mesh.xml")

# create a vector function space
V = dolfin.VectorFunctionSpace(mesh, "CG", 1)

# load the displacement field as a function
u = dolfin.Function(V)
with dolfin.HDF5File(dolfin.MPI.comm_world, "sample_data/displacements_final.h5", "r") as final_disp_file:
    final_disp_file.read(u, "data")
```

```python
dolfin.plot(u, title='Displacement Field')
```

```python
# load mesh
mesh = dolfin.Mesh("code/mesh.xml")

# create a vector function space
W = dolfin.FunctionSpace(mesh, "CG", 1)

# load the displacement field as a function
phi = dolfin.Function(W)
with dolfin.HDF5File(dolfin.MPI.comm_world, "sample_data/damage_final.h5", "r") as final_disp_file:
    final_disp_file.read(phi, "data")
```

```python
dolfin.plot(phi, title='Damage Field')
```
